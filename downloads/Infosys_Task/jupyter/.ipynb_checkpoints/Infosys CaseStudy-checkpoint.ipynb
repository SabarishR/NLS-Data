```python
# Feature Scaling
# there is no need to apply future scaling on dependent variable
sc = StandardScaler()
scalar = sc.fit(X_train)

# to save the scalar
joblib.dump(scalar,'../Model/scalar.pkl')

X = scalar.transform(X)
X_train = scalar.transform(X_train)
X_test = scalar.transform(X_test)
```

```python
# model optimal parameters
param = {
            "n_estimators" : 100,
            "max_depth" : 20,
            "min_samples_split" : 2,
            "min_samples_leaf" : 1,
            "bootstrap" : False,
            "random_state" : 42,
            "warm_start" : True
        }
```

```python
n = X.shape[0]
p = X.shape[1]
# calculating score values
print("r2 score: %.2f" % r2_score(y_test, y_pred))
print("explained variance score: %.2f" % explained_variance_score(y_test, y_pred))
```

```python
print("mae: %.2f" % mean_absolute_error(y_test, y_pred))
print("mse: %.2f" % mean_squared_error(y_test, y_pred))
print("rmse: %.2f" % mean_squared_error(y_test, y_pred,squared = False))
```

```python
# for text mining
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.sentiment import SentimentIntensityAnalyzer
from textblob import TextBlob
```

```python
# exploring the dataset
txt_df.head()
```

```python
# Missing value treatment
txt_df.info()
```

```python
# selecting the comments which is having values
txt_df = txt_df[txt_df['General Comment'].notna()].reset_index()
```

```python
# 1. Text cleaning
txt_df['General Comment']  = txt_df['General Comment'] .apply(lambda x: str(x).lower().strip())
```

```python
# Define function to remove punctuation and special characters
def remove_special_chars(text):
    return re.sub('[^A-Za-z0-9]+', ' ', text)
```

```python
# Apply function to 'text' column
txt_df['General Comment'] = txt_df['General Comment'].apply(remove_special_chars)
```

```python
# Define a function to remove stop words from a given string
def remove_stopwords(text):
    words = nltk.word_tokenize(text)
    filtered_words = [word.lower() for word in words if word.lower() not in stopwords.words('english')]
    return ' '.join(filtered_words)
```

```python
# Apply the remove_stopwords function to your column
txt_df['General Comment'] = txt_df['General Comment'].apply(remove_stopwords)
```

```python
# Stemming process
def stem_sentence(sentence):
    ps = PorterStemmer()
    words = word_tokenize(sentence)
    stemmed_words = [ps.stem(word) for word in words]
    return ' '.join(stemmed_words)
```

```python
txt_df['stemmed_sentences'] = txt_df['General Comment'].apply(stem_sentence)
```

```python
# Lammatization process
def lemmatize_sentence(sentence):
    lemmatizer = WordNetLemmatizer()
    words = sentence.split()
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    return ' '.join(lemmatized_words)
```

```python
txt_df['lemmatized_sentences'] = txt_df['General Comment'].apply(lemmatize_sentence)
```

```python
# Calculate overall sentiment using NLTK
sid = SentimentIntensityAnalyzer()
txt_df['sentiment'] = txt_df['General Comment'].apply(lambda x: sid.polarity_scores(x)['compound'])
```

```python
# Perform sentiment analysis using Textblob
txt_df['polarity'] =  txt_df['General Comment'].apply(lambda x: TextBlob(x).sentiment.polarity)
txt_df['subjectivity'] =  txt_df['General Comment'].apply(lambda x: TextBlob(x).sentiment.subjectivity)

# Print out the results
print(f"Average Polarity: {txt_df['polarity'].mean():.2f}")
print(f"Average Subjectivity: {txt_df['subjectivity'].mean():.2f}")
```

```python
# Filtering the sentiment having highly negative
df_sentiment = txt_df[(txt_df['sentiment'] <= -0.3) | (txt_df['polarity'] <= -0.3)]
df_sentiment['freq'] = df_sentiment.groupby(['Service Component'])['Service Component'].transform('count')
```

```python
# Selecting the area to be improved interms of service component
df_sentiment = df_sentiment[(df_sentiment['freq'] >= 5)]
print("Major improvement areas In terms of service- ",df_sentiment['Service Component'].unique())
```

```python
# Create a list of negative words using TextBlob polarity
neg_words = []
for text in txt_df['General Comment']:
    blob = TextBlob(text)
    for sentence in blob.sentences:
        if sentence.sentiment.polarity <= -0.2:
            for word in sentence.words:
                if word not in ['not', 'no']:
                    neg_words.append(word.lower())
```

```python
from wordcloud import WordCloud
# Create a word cloud of the negative words
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(neg_words))

# Plot the word cloud
plt.figure(figsize=(12, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
