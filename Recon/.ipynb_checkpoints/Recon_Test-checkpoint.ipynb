```python
# Query
query = """
select his.AC_ENTRY_SR_NO As GL_ID,his.ACCOUNT_NO AS GL_ACCNO,his.VALUE_DT as GL_DATE,his.DRCR_IND as GL_CRDR,
CASE 
    WHEN his.AC_CCY = 'KES' then his.LCY_AMOUNT
    ELSE his.FCY_AMOUNT
    END as GL_AMOUNT,
ft.USER_REF_NUMBER as GL_REF,ft.SOURCE_REF_NO GL_SOURCE_REF,
CASE
    WHEN ft.PAYMENT_DETAILS IS NOT NULL AND ft.PAYMENT_DETAILS != '' THEN ft.PAYMENT_DETAILS
    WHEN ct.ADDL_TEXT IS NOT NULL THEN ct.ADDL_TEXT
    ELSE null
    END as GL_ADDL_TEXT,
his.AC_CCY as GL_CURRENCY,recon.fld_UniqueReference as REFERENCE_ID
from statement$detail$history his 
left JOIN funds$transfer ft on ft.id = his.TRN_REF_NO
LEFT JOIN cstb$addl$text ct on ct.REFERENCE_NO = his.TRN_REF_NO
INNER JOIN tbl_recon_nostro$matched$entry recon on recon.PRI_entryId = his.AC_ENTRY_SR_NO
WHERE recon.fld_MatchType = 'MANUAL'
"""
```

```python
# Fetch GL data from DB
mysql_connection = mysql.connector.connect(**config)
GLdf = pd.read_sql(query, con=mysql_connection)
mysql_connection.close()
GLdf.tail()
```

```python
# Fetch NOSTRO data from DB
query = """
select ns.ID as RE_ID,ns.ACCOUNT_NUMBER as RE_ACCNO,ns.VALUE_DATE as RE_DATE,ns.CRF_TYPE as RE_CRDR,
ns.SWIFT_AMOUNT as RE_AMOUNT,ns.ACC_OWNER_REF_FOR_DEBIT,ns.ACC_SERVICING_INSTITUTE_REF_FOR_CREDIT, 
recon.fld_UniqueReference as REFERENCE_ID from nostro$statement ns
INNER JOIN tbl_recon_nostro$matched$entry recon on recon.PRI_entryId = ns.id
"""
mysql_connection = mysql.connector.connect(**config)
REdf = pd.read_sql(query, con=mysql_connection)
mysql_connection.close()
REdf.tail()
```

```python
# Converting time_stamp
GLdf['GL_DATE'] = pd.to_datetime(GLdf['GL_DATE'])
REdf['RE_DATE'] = pd.to_datetime(REdf['RE_DATE'])

# Remove negative signs
GLdf['GLAMOUNT'] = GLdf['GL_AMOUNT'].abs()
# RE_df['RE_AMOUNT'] = RE_df['RE_AMOUNT'].abs()

# Adding negative sign to Debit transactions
GLdf.loc[(GLdf['GL_CRDR'] == 'D'), 'GL_AMOUNT'] =  GLdf['GL_AMOUNT']*-1
# RE_df.loc[(RE_df['RE_CRDR'] == 'D'), 'RE_AMOUNT'] =  RE_df['RE_AMOUNT']*-1

# Replace CRDR value
REdf['RE_CRDR'] = REdf['RE_CRDR'].replace(['CREDIT', 'DEBIT'], ['C','D'])
REdf.head()
```

```python
# Define variance amount
Variance = 1
# Creating ismatched,match table and match id column
GLdf['ISMATCHED'] = 0
REdf['ISMATCHED'] = 0
GLdf['MATCH_TABLE'] = ''
GLdf['MATCH_ID'] = ''
# 
# Dataframe info
GLdf.info()
```

```python
REFERENCE_ID = GLdf['REFERENCE_ID'].unique().tolist()
Max_per_com = 3

for k in range(0,len(REFERENCE_ID)):
    print(k,REFERENCE_ID[k])
    GL_df = GLdf.loc[GLdf['REFERENCE_ID'] == str(REFERENCE_ID[k])]
    GL_df.reset_index(drop=True, inplace=True)
    RE_df = REdf.loc[REdf['REFERENCE_ID'] == str(REFERENCE_ID[k])]
    RE_df.reset_index(drop=True, inplace=True)
    
    # With in GL account match
    if(len(GL_df.loc[GL_df['ISMATCHED']==0]) > 1):
        for i in range(0,len(GL_df)-1):
            if (GL_df['ISMATCHED'][i] == 1):
                continue
            for j in range(i+1,len(GL_df)):
                if(GL_df['ISMATCHED'][j] == 1):
                    continue
                amount = abs(GL_df['GL_AMOUNT'][i] + GL_df['GL_AMOUNT'][j])
                if amount <= Variance:
                    GLdf.loc[GLdf['GL_ID'] == str(GL_df['GL_ID'][i]),'ISMATCHED'] = 1
                    GLdf.loc[GLdf['GL_ID'] == str(GL_df['GL_ID'][j]),'ISMATCHED'] = 1
                    GLdf.loc[GLdf['GL_ID'] == str(GL_df['GL_ID'][j]),'MATCH_TABLE'] = 'statement$detail$history'
                    GLdf.loc[GLdf['GL_ID'] == str(GL_df['GL_ID'][j]),'MATCH_ID'] = str(GL_df['GL_ID'][i])
                    break
    
    # One to one match with GL to Nostro to avoid more combinations
    for i in range(0,len(GL_df)):
        if (GL_df['ISMATCHED'][i] == 1):
            continue
        for j in range(0,len(RE_df)):
            if((RE_df['ISMATCHED'][j] == 1)):
                continue
            amount = abs(GL_df['GL_AMOUNT'][i] + RE_df['RE_AMOUNT'][j])
            if amount <= Variance:
                GLdf.at[GL_df.index[i],'ISMATCHED'] = 1
                REdf.at[RE_df.index[j],'ISMATCHED'] = 1
                GLdf.at[GL_df.index[i],'MATCH_TABLE'] = 'tbl_recon_nostro$matched$entry'
                GLdf.at[GL_df.index[i],'MATCH_ID'] = str(RE_df['RE_ID'][j])
                
                GLdf.loc[GLdf['GL_ID'] == str(GL_df['GL_ID'][i]),'ISMATCHED'] = 1
                REdf.loc[REdf['RE_ID'] == str(RE_df['RE_ID'][j]),'ISMATCHED'] = 1
                GLdf.loc[GLdf['GL_ID'] == str(GL_df['GL_ID'][i]),'MATCH_TABLE'] = 'tbl_recon_nostro$matched$entry'
                GLdf.loc[GLdf['GL_ID'] == str(GL_df['GL_ID'][i]),'MATCH_ID'] = str(GL_df['GL_ID'][i])
                break

    if((len(GL_df.loc[GL_df['ISMATCHED']==0])) > 0 & (len(RE_df.loc[RE_df['ISMATCHED']==0]) > 0)):

        dict_data = GL_df.loc[GL_df['ISMATCHED']==0].set_index('GL_ID')['GL_AMOUNT'].to_dict()
        length = len(dict_data)
        GL_per_com = [i for j in range(length + 1) for i in combinations(dict_data, j) if sum(map(dict_data.get, i))]

        GL_per_com_list = []
        for i in GL_per_com:
            if len(i) <= Max_per_com:
                GL_per_com_list.append(i)

        dict_data = RE_df.loc[RE_df['ISMATCHED']==0].set_index('RE_ID')['RE_AMOUNT'].to_dict()
        length = len(dict_data)
        RE_per_com = [i for j in range(length +  1) for i in combinations(dict_data, j) if sum(map(dict_data.get, i))]

        RE_per_com_list = []
        for i in RE_per_com:
            if len(i) <= Max_per_com:
                RE_per_com_list.append(i)

        for i in range(0,len(GL_per_com_list)):
            if 1 in (GL_df[GL_df['GL_ID'].isin(list(GL_per_com_list[i]))]['ISMATCHED'].values):
                continue
            if len(GL_df[GL_df['GL_ID'].isin(list(GL_per_com_list[i]))]['GL_DATE'].dt.strftime('%m/%Y').unique()) > 1:
                continue
            for j in range(0,len(RE_per_com_list)):
                if 1 in (RE_df[RE_df['RE_ID'].isin(list(RE_per_com_list[j]))]['ISMATCHED'].values):
                    continue
                if len(RE_df[RE_df['RE_ID'].isin(list(RE_per_com_list[j]))]['RE_DATE'].dt.strftime('%m/%Y').unique()) > 1:
                    continue
                if ((GL_df[GL_df['GL_ID'].isin(list(GL_per_com_list[i]))]['GL_DATE'].dt.strftime('%m/%Y').unique()[0]) != (RE_df[RE_df['RE_ID'].isin(list(RE_per_com_list[j]))]['RE_DATE'].dt.strftime('%m/%Y').unique()[0])):
                    continue
                amount = abs((GL_df[GL_df['GL_ID'].isin(list(GL_per_com_list[i]))]['GL_AMOUNT'].sum()) + (RE_df[RE_df['RE_ID'].isin(list(RE_per_com_list[j]))]['RE_AMOUNT'].sum()))
                if amount <= Variance:
                    GLdf.loc[GLdf['GL_ID'].isin(list(GL_per_com_list[i])),'ISMATCHED'] = 1
                    REdf.loc[REdf['RE_ID'].isin(list(RE_per_com_list[j])),'ISMATCHED'] = 1
                    GLdf.loc[GLdf['GL_ID'].isin(list(GL_per_com_list[i])),'MATCH_TABLE'] = 'tbl_recon_nostro$matched$entry'
                    GLdf.loc[GLdf['GL_ID'].isin(list(GL_per_com_list[i])),'MATCH_ID'] = str(list(RE_per_com_list[j]))
                    break
```
